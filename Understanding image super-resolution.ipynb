```python
# Read an the butterfly image and show its size
butterfly_image = "butterfly_GT.bmp"
image = scipy.misc.imread(butterfly_image, mode='RGB')
plt.imshow(image)
plt.show()
print('Size of image is: ' + str(image.shape))
```

```python
# Loading the image as YCbCr
image_lum = scipy.misc.imread(butterfly_image, mode='YCbCr', flatten=False)
plt.imshow(image_lum)
plt.show()
print('Size of image is: ' + str(image_lum.shape))
```
